EXERCISE 1:

Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

```python
def compute_pay(hours, rate):                       #function will calculate the pay based on the provided hours and rate
    if hours <= 40:                                 #if condition to check if the given hours are less than or equal to 40
        return hours * rate                         #return multiplying the hours and rate
    else:
        overtime_hours = hours - 40                 #calculates the number of overtime hours by subtracting 40 from the total hours
        overtime_pay = overtime_hours * (rate * 1.5)#calculates the overtime pay by multiplying the overtime hours with the rate multiplied by 1.5
        regular_pay = 40 * rate                     #calculates the regular pay by multiplying the regular hours (40) with the rate
        return regular_pay + overtime_pay           #returns the sum of regular pay and overtime pay as the total pay


hours = float(input("Enter Hours: "))               # Get the hours and rate from the user
rate = float(input("Enter Rate: "))

pay = compute_pay(hours, rate)                      # Calculate and print the pay
print(f"Pay: {pay}")
```

CHALLENGE

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

```python
def computegrade(score):                                                      #defines a function named computegrade that takes a score as its parameter
    grade_boundaries = [(97, 100), (93, 97), (89, 93), (85, 89), (81, 85), (77, 81), (74, 77), (0, 74)] #line creates a list of tuples named grade_boundaries
    grades = ["1.0", "1.25", "1.5", "1.75", "2.0", "2.25", "2.5", "3.0"]      # creates a list named grades

    for boundary, grade in zip(grade_boundaries, grades):                     #starts a loop that iterates through the grade_boundaries and grades lists
        if score >= boundary[0] and score < boundary[1]:                      #checks if the given score falls within the current grade boundary
            return grade                                                      #returns grade


    return "Invalid Score"                                                    #return a default grade if the score is not within any boundary


while True:                                                                   #test the function with different scores
    score = float(input("Enter a score "))
    if score == 'exit':                                                       #checks if the entered score is equal to 'exit'
        break                                                                 #If the user enters 'exit', this line breaks the loop
    grade = computegrade(score)                                               #if the score is not 'exit  w' this line calls the computegrade function with the entered score and assigns the returned grade to the grade variable
    print(f"Grade for score {score}: {grade}")                                #prints the calculated grade for the given score in a formatted string
```
