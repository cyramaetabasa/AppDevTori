Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

```python
hrs = float(input("Enter Hours: ")) #user enter hours
rate = float(input("Enter Rate: ")) # user enter rate

if hrs >= 40:
    pay = (40 * rate) + ((hrs - 40) * rate * 1.5) #if hours is greater than or equal to 40 calculate pay with overtime hours at 1.5 times the rate
else:
    pay == hrs * rate
     #if its not greater than or equal to 40 hours calculate pay for all hours at the regular rate

print("Pay:", pay) #display calculated pay
```

CHALLENGE

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

```python
def grade_for_score(score):
    if score >= 97:#if condition to check if the score is greater than or equal to 97
        return 1.0
    elif score >= 93:#else-if condition to check if the score is greater than or equal to 93
        return 1.25
    elif score >= 89:# else-if condition to check if the score is greater than or equal to 89
        return 1.5
    elif score >= 85:# else-if condition to check if the score is greater than or equal to 85
        return 1.75
    elif score >= 81:#else-if condition to check if the score is greater than or equal to 81
        return 2.0
    elif score >= 77:#else-if condition to check if the score is greater than or equal to 77
        return 2.25
    elif score >= 74:#else-if condition to check if the score is greater than or equal to 74.
        return 2.5
    else:#if none of the previous conditions were true
        return 3.0


score = float(input("Enter a score between 0 and 100: "))#Get the score from the user


grade = grade_for_score(score)#Calculate and print the grade
print(f"Grade: {grade}")
```
