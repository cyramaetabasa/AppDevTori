Write a program which repeatedly reads numbers until the user enters "done". Once "done" is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

```python
total = 0                               #ines initialize the variables total and count to 0
count = 0

while True:                             #starts an infinite loop that will continue until the user enters "done"
    try:
        num = input("Enter a number: ") #user to enter a number, and the input is stored in the variable num

        if num.lower() == 'done':       # if the user entered "done
            break

        number = float(num)             #if successful the input is considered a valid number, and it is stored in the variable number
        total += number                 #update the total and count variables by adding the current number to total and incrementing count by 1
        count += 1

    except ValueError:
        print("Invalid input")
        continue

if count > 0:                          #if there are any valid numbers entered
    average = total / count            #calculates the average of the valid numbers by dividing the total by the count
    print(f"{total} {count} {average:.3f}")#prints the total, count, and average in the specified format
else:
    print(f"You didn't enter any numbers.")#prints this line if your did not enter number
```
