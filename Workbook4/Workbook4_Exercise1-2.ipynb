- Create your own module.
- Place variables and functions inside it.
- Import your module and use the variabls and functions you've created in the module.
- Try to add docstrings for the module and the functions

```python
"""A module containing variables and functions."""

# Module-level variable
MODULE_CONSTANT = 42

def greet(name):
    """Greet the user by name."""
    return f"Hello, {name}!"

def add(x, y):
    """Add two numbers."""
    return x + y


# Use the module-level variable
print("Module constant:",MODULE_CONSTANT)

# Use the function in the module
print(greet("Alice"))

# Use another function in the module
result = add(3, 5)
print("Addition result:", result)
```
